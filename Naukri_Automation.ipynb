```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

email = "" #Enter your authorized mail ID 
password = "" #Enter your Password
job_keyword = "Data Analyst"
location = "Bengaluru"
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    driver.get("https://www.naukri.com/")
    time.sleep(3)
    login_button = driver.find_element(By.LINK_TEXT, "Login")
    login_button.click()
    time.sleep(3)
    email_field = driver.find_element(By.XPATH, '//input[@placeholder="Enter your active Email ID / Username"]')
    email_field.send_keys(email)
    password_field = driver.find_element(By.XPATH, '//input[@type="password"]')
    password_field.send_keys(password)
    driver.find_element(By.CLASS_NAME, "btn-primary").click()
    time.sleep(5)
    search_box = driver.find_element(By.XPATH, '//span[contains(@class, "nI-gNb-sb__placeholder")]')
    search_box.click()
    time.sleep(1)
    search_input = driver.find_element(By.XPATH, '//input[@type="text"]')
    search_input.send_keys(job_keyword)
    time.sleep(1)
    location_box = driver.find_element(By.XPATH, '//input[contains(@class, "suggestor-input") and @placeholder="Enter location"]')
    location_box.send_keys(location)
    time.sleep(1)
    search_button = driver.find_element(By.XPATH, '//button[contains(@class, "nI-gNb-sb__icon-wrapper")]')
    search_button.click()
    time.sleep(5)
    
    try:

        sort_dropdown = driver.find_element(By.XPATH, '//button[contains(@id, "filter-sort")]')
        sort_dropdown.click()
        time.sleep(2)

     
        recommended_filter = driver.find_element(By.XPATH, '//li[contains(@title, "Recommended")]')
        recommended_filter.click()
        time.sleep(2)
    
        sort_dropdown.click()
        time.sleep(2)
        date_filter = driver.find_element(By.XPATH, '//li[contains(@title, "Date")]')
        date_filter.click()
        time.sleep(2)
    except Exception as e:
        print(f"‚ö†Ô∏è Failed to apply filters: {e}")
    jobs = driver.find_elements(By.XPATH, '//div[contains(@data-job-id, "")]')
    job_index = 0

    while job_index < len(jobs):
        try:
            job = jobs[job_index]
            job.click()
            time.sleep(2)
            driver.switch_to.window(driver.window_handles[-1])
            time.sleep(2)

            if "Applied" in driver.page_source or "Apply on company site" in driver.page_source:
                print("‚è© Job already applied or redirects to external site. Skipping...")
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                job_index += 1
                continue



            
            apply_button = driver.find_element(By.XPATH, '//button[contains(@class, "apply-button") and @id="apply-button"]')
            apply_button.click()
            time.sleep(2)
            print("‚úÖ Applied to a job!")

            
            driver.close()
            driver.switch_to.window(driver.window_handles[0])

        except Exception as e:
            print(f"‚ùó Failed to apply to this job: {e}")
            driver.switch_to.window(driver.window_handles[0])

        job_index += 1
        time.sleep(2)

        
        jobs = driver.find_elements(By.XPATH, '//div[contains(@data-job-id, "")]')

    print("üéâ Done applying to jobs!")

except Exception as e:
    print(f"Error: {e}")

finally:
    driver.quit()
```
